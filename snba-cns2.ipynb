- **Handling Negative values**
- We'll flag negative values and multiply them by -1 to preserve the magnitude

```python
count_df = count_df.copy()
cols_to_fix = [c for c in count_df.columns if count_df[c].dtype in ['int64', 'float64'] and
              count_df[c].min() < 0]

print (f'Columns with negative values: {cols_to_fix}')

for col in cols_to_fix:
    count_df[f'{col}_was_negative'] = (count_df[col] < 0).astype(bool)
    count_df[col] = np.abs(count_df[col])
```

- handle `inf` values for `Flow Bytes/s`

```python
max_finite = count_df[count_df['Flow Bytes/s'] != np.inf]['Flow Bytes/s'].max()
count_df['Flow Bytes/s_had_inf'] = count_df['Flow Bytes/s'] == np.inf
count_df['Flow Bytes/s'] = count_df['Flow Bytes/s'].replace(np.inf, max_finite)

print(f"Inf values present: {(count_df['Flow Bytes/s'] == float('inf')).any()}")
```

- Applying `logp1` transformation

```python
count_df['Flow Bytes/s'] = np.log1p(count_df['Flow Bytes/s'])
count_df['Flow Bytes/s'].describe()
```

- Fill null values with median

```python
count_df['Flow Bytes/s_was_missing'] = count_df['Flow Bytes/s'].isna().astype(bool)

median_val = count_df['Flow Bytes/s'].median()
count_df['Flow Bytes/s'] = count_df['Flow Bytes/s'].fillna(median_val)

print(f"Remaining nulls: {count_df['Flow Bytes/s'].isna().sum()}")
```

- Get a list of skewed features, and a list of normally distributed features

```python
from scipy.stats import skew

right_skewed = []
left_skewed = []
normal_ish = []

for col in count_df.columns:
    if count_df[col].dtype not in ['int', 'float']:
        continue

    data = count_df[col].dropna()
    if len(data) > 0:
        skew_val = skew(data)

        if skew_val > 1.0:
            right_skewed.append(col)
        elif skew_val < -1.0:
            left_skewed.append(col)
        else:
            normal_ish.append(col)

print(f"Right-Skewed (Log it): {len(right_skewed)}")
print(f"Left-Skewed (Square it): {len(left_skewed)}")
print(f"Normal-ish: {len(normal_ish)}")
```

```python
for c in right_skewed:
  print (f'{c:<30} has {len(count_df[c].unique())} uniques')
```

- Notice that `Flow Bytes/s` became normally distributed after transformation and imputation, and so the case with all the right-skewed features
- Next, In order to handle the outliers the remanining features, we apply the same steps that we've done to `Flow Bytes/s` to all the remanining right-skewed descriptive features.
- And check whether their ourtliers are now gone or still.
- As for `NaN` values, it has been shown that none remain, same for `inf`.

```python
inf_columns = [c for c in count_df.columns if (count_df[c] == -np.inf).any() or (count_df[c] == np.inf).any()]

for col in inf_columns:
    # Create flag for each inconsistent feature (having inf values)
    count_df[f'{col}_had_inf'] = ((count_df[col] == np.inf) | (count_df[col] == -np.inf)).astype(bool)

    max_finite = count_df[count_df[col] != np.inf][col].max()
    min_finite = count_df[count_df[col] != -np.inf][col].min()

    count_df[col] = count_df[col].replace(np.inf, max_finite)
    count_df[col] = count_df[col].replace(-np.inf, min_finite)
```

```python
from scipy.stats import skew

print("\n" + "=" * 120)
print(f"{'Feature':<30} | {'Unique Before':<12} | {'Unique After':<12} | {'Raw Skew':<10} | {'Final Skew':<10} | {'Improvement':<10}")
print("-" * 120)

for col in right_skewed:
    before_skew = skew(count_df[col])
    before_unique = count_df[col].nunique()

    # Apply log1p
    transformed_data = np.log1p(count_df[col])

    after_skew = skew(transformed_data)
    after_unique = len(np.unique(transformed_data))
    reduction = abs(before_skew) - abs(after_skew)

    print(f"{col:<30} | {before_unique:<12} | {after_unique:<12} | {before_skew:<10.4f} | {after_skew:<10.4f} | {reduction:<10.4f}")
```

- Generally, the `log1p` transformation was good enough in reducing the skewness of the data and its tendency towards its outliers.

### Correlation

- We'll plot a heatmap to see the correlation between each pair of features.
- But first, the `Label` feature's categorical values need to be handled, and `Label Encoding` is generally adequate for the label variable, as it won't be used by ML algorithms, but instead predicted, so the imposed numerical order after encoding (`label1 -> 0`, `label2 -> 1`, etc.) wouldn't impact a thing.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

# Fit and transform the Label column
count_df['Label'] = le.fit_transform(count_df['Label'])

print("Label mapping:")
for i, label in enumerate(le.classes_):
    print(f"  {label} -> {i}")
```

```python
matrix = count_df.drop(columns=['Label']).corr()

plt.figure(figsize=(20, 20))
sns.heatmap(matrix, 
            cmap='coolwarm', 
            center=0,
            annot=False,
            square=True,
            linewidths=0.5,
            cbar_kws={"shrink": 0.8})

# Rotate labels for better visibility
plt.xticks(rotation=90, fontsize=8)
plt.yticks(rotation=0, fontsize=8)

plt.title('Features Correlation Heatmap', fontsize=16)
plt.tight_layout()
plt.show()
```

- It's noticeable that some features are highly correlated.
- We'll extract highly correlated pairs, and for each pair, we'll calculate the correlation of both feature with the label feature `Label`, then drop the weaker correlated feature.

```python
target_corr = count_df.drop(columns=['Label']).corrwith(count_df['Label']).abs()

# Find highly correlated pairs
threshold = 0.9
upper_tri = matrix.where(np.triu(np.ones(matrix.shape), k=1).astype(bool))
high_corr_pairs = [(col, row, upper_tri.loc[row, col]) 
                   for col in upper_tri.columns 
                   for row in upper_tri.index 
                   if abs(upper_tri.loc[row, col]) > threshold]

to_drop = set()
for col1, col2, corr in high_corr_pairs:
    print(f"\nPair: {col1} <-> {col2} (correlation: {corr:.4f})")
    print(f"  {col1} vs Label: {target_corr[col1]:.4f}")
    print(f"  {col2} vs Label: {target_corr[col2]:.4f}")
    print(f"  -> Keeping: {col1 if target_corr[col1] >= target_corr[col2] else col2}")
    if target_corr[col1] < target_corr[col2]:
        to_drop.add(col1)
    else:
        to_drop.add(col2)

keep_cols = [col for col in count_df.columns if col not in to_drop]
print(f"Dropped {len(to_drop)} features, kept {len(keep_cols)}")
```

```python
print("\n" + "="*60)
print("DROPPED FEATURES (with correlation to Label):")
print("="*60)
for col in sorted(to_drop):
    print(f"  {col:<35} | {target_corr[col]:.4f}")

print("\n" + "="*60)
print("KEPT FEATURES (with correlation to Label):")
print("="*60)
for col in sorted(keep_cols):
    if col == 'Label':
        continue
    print(f"  {col:<35} | {target_corr[col]:.4f}")
```

## SMOTE
- We'll apply `SMOTE` technique to reduce the imbalancy between the `Label` feature values seen previously.

```python
X_train.shape, y_train.shape
```

- With this data size, `SMOTE` would be very computationaly heavy, therefore, The number of instances of the majority class `BENIGN` need to be reduced before applying `SMOTE`, known as `Undersampling`.

```python
# Get class counts
from collections import Counter
class_counts = Counter(y_train)

# Find the largest minority class (excluding BENIGN)
minority_counts = {k: v for k, v in class_counts.items() if k != 'BENIGN'}
largest_minority = max(minority_counts.values())

print (f'Largest minority class count: {largest_minority}')
```

- classes are actually encoded, therefore, we need to find the encoding of `BENIGN`

```python
# Find BENIGN's encoded value
benign_encoded = le.transform(['BENIGN'])[0]
print(f"BENIGN encoded as: {benign_encoded}")
```

```python
from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline

# Set target size for BENIGN (e.g., 20% of largest minority)
benign_target = int(largest_minority * 0.2)

under_strategy = {benign_encoded: benign_target}

resample_pipeline = Pipeline([
    ('undersample', RandomUnderSampler(sampling_strategy=under_strategy, random_state=42)),
    ('smote', SMOTE(random_state=42))
])

X_train_resampled, y_train_resampled = resample_pipeline.fit_resample(X_train, y_train)

print(f"Original: {X_train.shape[0]} Training rows")
print(f"Resampled: {X_train_resampled.shape[0]} training rows")
```
